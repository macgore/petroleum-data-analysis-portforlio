```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

print("Week 2 enviroment ready")
```

```python
data = {
    "Month": pd.date_range(start="2023-01-01", periods=12, freq="M"),
    "Oil_Rate_bopd": [1600, 1550, 1500, 1420, 1350, 1280, 1200, 1140, 1080, 1020, 970, 920],
    "Gas_Rate_mscfd": [950, 940, 920, 900, 880, 860, 840, 820, 800, 780, 760, 740],
    "Water_Cut": [0.18, 0.20, 0.22, 0.25, 0.28, 0.30, 0.34, 0.38, 0.42, 0.46, 0.50, 0.55]
}

df = pd.DataFrame(data)
df
```

```python
df.head()
```

```python
df.describe()
```

```python
df["Oil_Produced_bbl"] = df["Oil_Rate_bopd"] * 30
df["Gas_Produced_mscf"] = df["Gas_Rate_mscfd"] * 30

df.head()
```

```python
df["Cumulative_Oil_bbl"] = df["Oil_Produced_bbl"].cumsum()
df["Cumulative_Gas_mscf"] = df["Gas_Produced_mscf"].cumsum()

df.tail()
```

```python
df["GOR_scf_bbl"] = (df["Gas_Rate_mscfd"] * 1000) / df["Oil_Rate_bopd"]
df["Water_Cut_percent"] = df["Water_Cut"] * 100

df.head()
```

```python
df["Operational_Flag"] = np.where(df["Water_Cut"] > 0.45, "High Water Cut", "Normal")

df[["Month", "Water_Cut", "Operational_Flag"]]
```

```python
df.to_csv("well_production_cleaned.csv", index=False)

print("File saved: well_production_cleaned.csv")
```
